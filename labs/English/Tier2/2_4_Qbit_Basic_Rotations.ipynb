- `Rz(θ)` → rotation around the Z-axis by angle θ  

Each takes **one argument**, the **rotation angle θ**, measured in **radians**.

## 3. Explain — Conceptual and Mathematical Modeling
> **Purpose:** Translate geometric understanding into the formal representation of quantum rotation.

**Key concept:**  
A rotation changes the *phase* and/or *amplitude* of a qubit’s state depending on the chosen axis.  
Each rotation can be described by a **unitary matrix** that preserves normalization.

---

### 🔹 Mathematical formulation

The rotation operators are defined as:

$$
R_x(θ) =
\begin{bmatrix}
\cos\left(\frac{θ}{2}\right) & -i\sin\left(\frac{θ}{2}\right) \\
-i\sin\left(\frac{θ}{2}\right) & \cos\left(\frac{θ}{2}\right)
\end{bmatrix}
$$

$$
R_y(θ) =
\begin{bmatrix}
\cos\left(\frac{θ}{2}\right) & -\sin\left(\frac{θ}{2}\right) \\
\sin\left(\frac{θ}{2}\right) & \cos\left(\frac{θ}{2}\right)
\end{bmatrix}
$$

$$
R_z(θ) =
\begin{bmatrix}
e^{-iθ/2} & 0 \\
0 & e^{iθ/2}
\end{bmatrix}
$$

Each operator rotates the qubit around the **X**, **Y**, or **Z** axis of the Bloch sphere.

---

### 🔹 Example 1 — Rotation around the X-axis

Let the initial state be:

$$
|ψ⟩ = |0⟩ =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

Apply \( R_x\left(\frac{π}{2}\right) \):

$$
R_x\left(\frac{π}{2}\right)
=
\begin{bmatrix}
\cos(π/4) & -i\sin(π/4) \\
-i\sin(π/4) & \cos(π/4)
\end{bmatrix}
=
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 & -i \\
-i & 1
\end{bmatrix}
$$

Then,

$$
|ψ'⟩ = R_x\left(\frac{π}{2}\right)|0⟩ =
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 \\
-i
\end{bmatrix}
$$

This corresponds to a **rotation of 90° around the X-axis**, moving the state from the north pole toward the equator of the Bloch sphere.  
The probabilities are still normalized:

$$
P(0) = |1/\sqrt{2}|^2 = 0.5, \quad P(1) = |-i/\sqrt{2}|^2 = 0.5
$$

---

### 🔹 Example 2 — Rotation around the Y-axis

Starting again from \( |ψ⟩ = |0⟩ \):

$$
R_y\left(\frac{π}{2}\right)
=
\begin{bmatrix}
\cos(π/4) & -\sin(π/4) \\
\sin(π/4) & \cos(π/4)
\end{bmatrix}
=
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 & -1 \\
1 & 1
\end{bmatrix}
$$

Apply to \( |0⟩ \):

$$
|ψ'⟩ = R_y\left(\frac{π}{2}\right)|0⟩ =
\frac{1}{\sqrt{2}}
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

Now the qubit points along the **X-axis** of the Bloch sphere —  
a perfect **superposition state** with equal probability:

$$
P(0) = 0.5, \quad P(1) = 0.5
$$

and no relative phase.

---

### 🔹 Example 3 — Rotation around the Z-axis

Let’s start from a balanced superposition:

$$
|ψ⟩ = \frac{1}{\sqrt{2}}
\begin{bmatrix}
1 \\
1
\end{bmatrix}
$$

Apply  $  R_z\left(\frac{π}{2}\right) $:

$$
R_z\left(\frac{π}{2}\right)
=
\begin{bmatrix}
e^{-iπ/4} & 0 \\
0 & e^{iπ/4}
\end{bmatrix}
$$

Then,

$$
|ψ'⟩ =
\frac{1}{\sqrt{2}}
\begin{bmatrix}
e^{-iπ/4} \\
e^{iπ/4}
\end{bmatrix}
$$

The **probabilities remain unchanged**:

$$
P(0) = |e^{-iπ/4}/\sqrt{2}|^2 = 0.5, \quad P(1) = |e^{iπ/4}/\sqrt{2}|^2 = 0.5
$$

but the **relative phase** between ∣0⟩ and ∣1⟩ is now **π/2**, changing the qubit’s orientation on the equator of the Bloch sphere.

---

**Physical interpretation:**  
> - \(R_x(θ)\) and \(R_y(θ)\) change both amplitude and phase — they move the state through the Bloch sphere.  
> - \(R_z(θ)\) changes only the phase — it rotates the state *around* the Z-axis, without altering measurement probabilities.  
> - All rotations are **unitary**: normalization is preserved, ensuring total probability = 1.


## 4. Elaborate — Computational Exploration (Qiskit SDK)
> **Purpose:** Apply the concept of rotation using Qiskit to visualize how each rotation acts on the qubit.

```python
# Qiskit code — Rotation and Bloch Visualization
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import numpy as np

# Example angles
theta_x = np.pi / 2
theta_y = np.pi / 3
theta_z = np.pi / 4

# Create a single qubit circuit
qc = QuantumCircuit(1)

# Apply rotations
qc.rx(theta_x, 0)
qc.ry(theta_y, 0)
qc.rz(theta_z, 0)

# Get the final statevector
state = Statevector.from_instruction(qc)

# Display circuit and Bloch vector
display(qc.draw('mpl'))
plot_bloch_multivector(state)


````python
# Qiskit code — Rotation and Bloch Visualization
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import numpy as np

# Example angles
theta_x = np.pi / 2
theta_y = np.pi / 3
theta_z = np.pi / 4

# Create a single qubit circuit
qc = QuantumCircuit(1)

# Apply rotations
qc.rx(theta_x, 0)
qc.ry(theta_y, 0)
qc.rz(theta_z, 0)

# Get the final statevector
state = Statevector.from_instruction(qc)

# Display circuit and Bloch vector
display(qc.draw('mpl'))
plot_bloch_multivector(state)
````
