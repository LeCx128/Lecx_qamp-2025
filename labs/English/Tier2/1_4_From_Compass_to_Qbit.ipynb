## 2. Explore — Hands-On Activity: Paper Strip and Phase Projection

**Materials:**  
Paper strip (10–15 cm), tape, flashlight or lamp, whiteboard or wall.

**Procedure:**  
1. Attach the strip vertically to the wall or board.  
2. Create a Cartesian plane using the base of the strip as the center.  
3. Draw a circle using the length of the paper strip as the diameter.  
4. Illuminate it perpendicularly with the flashlight: observe the shadow.  
5. Rotate the strip a few degrees in the air → the shadow changes position.  
6. Move the flashlight (without rotating the strip) → the shadow changes even though the physical orientation remains the same.  

**Purpose:**  
- Design a model corresponding to the different configurations of the magnetic field.  
- Introduce the idea of a **non-directly visible simultaneity** (superposition).  
- Connect that orientation with the **mathematical description of the state**.

## 3. Explain — From the Physical Compass to the Quantum State ∣ψ⟩ = α∣0⟩ + β∣1⟩

**Objective:**  
Formalize the physical experience with the paper strip and project it to the mathematical plane: modeling the three-dimensional state vector.

**Step-by-Step (Instructor Guide)**

### 3.1 From Poles to the Z Axis

In the vertical plane of the board, identify two extreme states:

- When the needle points fully toward **north** (up along the Z-axis):  
  → we call it state **∣0⟩**.  
- When it points fully toward **south** (down along the Z-axis):  
  → it will be state **∣1⟩**.

These two states form our **computational basis**, i.e., the fundamental reference points.

### 3.2 A Compass Between Two Worlds

If we tilt the needle, it is no longer fully in ∣0⟩ or ∣1⟩.  
We can describe its intermediate position with an **angle θ**, measured from the Z-axis.  
The projection onto **Z+** corresponds to the **probability of being in ∣0⟩**, and the projection onto **Z−** to the **probability of being in ∣1⟩**.

Thus:

$$
P(0) = \cos^2\left(\frac{θ}{2}\right), \qquad
P(1) = \sin^2\left(\frac{θ}{2}\right)
$$

These probabilities always sum to 1.

### 3.3 From Angles to Complex Amplitudes

The complete orientation of the compass depends not only on the angle θ, but also on the rotation around the Z-axis, which we call **phase φ**.  
This rotation does not change the inclination, but it does change the “internal rotation” of the vector.

Combining both parameters, we can write the quantum state as:

$$
∣ψ⟩ = \cos\left(\frac{θ}{2}\right)∣0⟩ + e^{iφ}\sin\left(\frac{θ}{2}\right)∣1⟩
$$

Here:  
- $α = \cos\left(\frac{θ}{2}\right)$  
- $β = e^{iφ}\sin\left(\frac{θ}{2}\right)$  

And they satisfy the fundamental condition:  
$$|α|^2 + |β|^2 = 1.$$

### 4.4 Physical Visualization Strategies  
> Use the **Bloch hemisphere** you built in the Engage activity.  
> Invite students to move the paper strip representing θ and φ.  
> Mark the equatorial plane (maximum superposition) with one color and the Z-axis (pure states ∣0⟩ and ∣1⟩) with another.  
> The goal is to visualize how a physical orientation translates into a **probabilistic and complex combination**.


```python
from qiskit import QuantumCircuit
from qiskit.visualization import plot_bloch_multivector
from qiskit.quantum_info import Statevector
import numpy as np

# Orientation parameters
theta = np.pi / 3   # Vector inclination (60°)
phi = np.pi / 4     # Phase or rotation around Z-axis (45°)

# Single-qubit circuit
qc = QuantumCircuit(1)
qc.ry(theta, 0)     # Y rotation -> controls angle θ
qc.rz(phi, 0)       # Z rotation -> controls phase φ
qc.draw('mpl')

# Get the state vector
state = Statevector.from_instruction(qc)
display(plot_bloch_multivector(state))
```

### 4.2 Guided Exploration

> - Ask the following questions to the teacher trainees:

>     *(Hint: the vector points to the north or south pole of the sphere, representing ∣0⟩ or ∣1⟩).*  
>   - What happens if we change φ while keeping θ fixed?  
>     *(Hint: the vector rotates around the Z-axis without changing measurement probabilities, only its relative phase).*

> This dialogue helps teachers recognize **quantum phase** as a property that does not change probabilities, but affects the **coherence** of the state.

### 4.3 Conceptual Connection

The Qiskit simulation illustrates that a **pure quantum state** can be written as:  

$$
∣ψ⟩ = \cos\left(\frac{θ}{2}\right)∣0⟩ + e^{iφ}\sin\left(\frac{θ}{2}\right)∣1⟩
$$  

where:
- $θ$ controls the **vector inclination** (analogous to the compass angle).  
- $φ$ represents the **phase** or rotation around the Z-axis.  

> **Pedagogical Interpretation:**  
> - $|α|^2 = \cos^2\left(\frac{θ}{2}\right)$ → probability of finding the qubit in state ∣0⟩.  
> - $|β|^2 = \sin^2\left(\frac{θ}{2}\right)$ → probability of finding it in ∣1⟩.  

> **Instructor Suggestion:**  
> Draw a Bloch hemisphere on the board with tape or string, and show how a change in $θ$ and $φ$ “moves” the vector ∣ψ⟩.  
> This helps participants visualize the connection between **physical rotations** and **unitary transformations**.
