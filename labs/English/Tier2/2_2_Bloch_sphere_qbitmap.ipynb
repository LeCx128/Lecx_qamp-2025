## 2. Explore — Discover the Tools

> In this phase, we’ll get to know the tools that Qiskit provides to visualize and analyze a qubit’s state — without coding yet.

### 2.1 Main Libraries

To represent and visualize the quantum state, we’ll rely on three key components:

1. **`QuantumCircuit` (from Qiskit):**  
   - Builds a circuit with one or more qubits.  
   - Each qubit can be modified using rotations.  
   - Common arguments:
     - `QuantumCircuit(n)` → creates a circuit with `n` qubits.
     - `.rx(θ, q)` → rotates qubit `q` by angle θ around the Y-axis.
     - `.ry(θ, q)` → rotates qubit `q` by angle θ around the Y-axis.
     - `.rz(φ, q)` → rotates qubit `q` by angle φ around the Z-axis.

2. **`Statevector` (from `qiskit.quantum_info`):**  
   - Represents the quantum state of the circuit.  
   - Allows calculation of amplitudes and measurement probabilities.  
   - Created using:
     - `Statevector.from_instruction(circuit)`

3. **`plot_bloch_multivector` (from `qiskit.visualization`):**  
   - Draws the qubit state on the Bloch sphere.  
   - Displays its position according to the angles θ (inclination) and φ (phase rotation).

---

### 2.2 Fundamental Variables

| Variable | Meaning | Typical Range | Effect |
|-----------|----------|---------------|--------|
| θ (theta) | Inclination angle | 0 to π | Controls probability balance between ∣0⟩ and ∣1⟩ |
| φ (phi) | Phase angle | 0 to 2π | Controls orientation in the XY plane |

> :D Remember *θ defines the mix between ∣0⟩ and ∣1⟩, while φ defines the orientation of that mix.*

## 3. Explain — From Angles to the Quantum State

The general quantum state can be written as:

$$
∣ψ⟩ = \cos\left(\frac{θ}{2}\right)∣0⟩ + e^{iφ}\sin\left(\frac{θ}{2}\right)∣1⟩
$$

---

### 3.1 Geometric Representation

On the Bloch sphere:
- The **Z-axis** separates the classical states ∣0⟩ (north) and ∣1⟩ (south).  
- The angle **θ** controls the height (probability mixture).  
- The angle **φ** controls the rotation around the Z-axis (phase).

The vector form of the state is:

$$
\vec{r} =
\begin{pmatrix}
\sin θ \cos φ \\
\sin θ \sin φ \\
\cos θ
\end{pmatrix}
$$

> - When θ = 0 → the qubit is fully ∣0⟩  
> - When θ = π → the qubit is fully ∣1⟩  
> - When θ = π/2 → the qubit is in an equal superposition  

---

### 3.2 Relation to Probabilities

| θ | φ | P(0) | P(1) | Interpretation |
|----|----|------|------|----------------|
| 0 | 0 | 1.0 | 0.0 | Pure ∣0⟩ |
| π/2 | 0 | 0.5 | 0.5 | Real superposition |
| π/2 | π/2 | 0.5 | 0.5 | Superposition with phase |
| π | 0 | 0.0 | 1.0 | Pure ∣1⟩ |

> The value of φ does not change probabilities — it only rotates the vector in the XY plane.

## 4. Elaborate — Apply and Visualize with Qiskit

> Now let’s put everything together by creating a Qiskit circuit and visualizing the state on the Bloch sphere.

### 4.1 Visualizing the Quantum State

```python
from qiskit.circuit import QuantumCircuit
from qiskit.visualization import plot_bloch_vector
import numpy as np
import matplotlib.pyplot as plt

# --- PARÁMETROS ---
theta = np.deg2rad(45)  # 60 grados → radianes
phi = np.deg2rad(45)    # 45 grados → radianes

# You can use spherical coordinates instead of cartesian.
plot_bloch_vector([1, theta, phi], coord_type='spherical')

## 4.2 — State ∣0⟩ (θ = 0, φ = 0)


````python

from qiskit.visualization.bloch import Bloch
import numpy as np
import matplotlib.pyplot as plt

# Parámetros
theta = 0
phi = 0

# Coordenadas Bloch
x = np.sin(theta) * np.cos(phi)
y = np.sin(theta) * np.sin(phi)
z = np.cos(theta)

# Crear esfera
b = Bloch()
b.add_vectors([x, y, z])
b.render()
plt.title("|0>  →  θ=0, φ=0", fontsize=12)
plt.show()
````

## 4.3 State "Superposition"  (θ = π/2, φ = 0)

````python
from qiskit.visualization.bloch import Bloch
import numpy as np
import matplotlib.pyplot as plt

theta = np.pi/2 # 90° 
phi = 0

x = np.sin(theta) * np.cos(phi)
y = np.sin(theta) * np.sin(phi)
z = np.cos(theta)

b = Bloch()
b.add_vectors([x, y, z])
b.render()
plt.title("Superposition  →  θ=π/2, φ=0", fontsize=12)
plt.show()
````

## 4.4 "Phased Superposition"  (θ = π/2, φ = π/2)

````python
from qiskit.visualization.bloch import Bloch
import numpy as np
import matplotlib.pyplot as plt

theta = np.pi/2
phi = np.pi/2

x = np.sin(theta) * np.cos(phi)
y = np.sin(theta) * np.sin(phi)
z = np.cos(theta)

b = Bloch()
b.add_vectors([x, y, z])
b.render()
plt.title("Phased Superposition  →  θ=π/2, φ=π/2", fontsize=12)
plt.show()
````

## 4.5 State ∣1⟩ (θ = π, φ = 0)

````python
from qiskit.visualization.bloch import Bloch
import numpy as np
import matplotlib.pyplot as plt

theta = np.pi
phi = 0

x = np.sin(theta) * np.cos(phi)
y = np.sin(theta) * np.sin(phi)
z = np.cos(theta)

b = Bloch()
b.add_vectors([x, y, z])
b.render()
plt.title("|1>  →  θ=π, φ=0", fontsize=12)
plt.show()
````

## 4.6 Mixed Quantum State (45° , 45°) First full Quantum Simulation

### 4.6.1 Set de angles for Theta= 45 and for Phi = 45

````python
from qiskit.circuit import QuantumCircuit
import numpy as np
import matplotlib.pyplot as plt

# --- PARÁMETROS ---
theta = np.deg2rad(45)  # 60 grados → radian
phi = np.deg2rad(45)    # 45 grados → radian

# Crear circuito
qc = QuantumCircuit(1)
qc.ry(theta, 0)   # rotation Y -  declination -  vertical
qc.rz(phi, 0)     # rotation Z (phase) - Horizontal
qc.measure_all()
qc.draw("mpl")

````

### 4.6.2 Visualize the expected measurement of angles Theta and Phi on Bloch Sphere

````python
import numpy as np
from qiskit.visualization import plot_bloch_vector

# You can use spherical coordinates instead of cartesian.
plot_bloch_vector([1, theta, phi], coord_type='spherical')
````

### 4.6.3 Measure the Probabily od being in `|0>`   North Pole  and South Pole `|1>`  

````python
# Add measurement and simulate
from qiskit_aer.primitives import SamplerV2
sampler = SamplerV2()
job = sampler.run([qc], shots=128)

# Perform an ideal simulation
result_ideal = job.result()
counts_ideal = result_ideal[0].data.meas.get_counts()
print('Counts(ideal):', counts_ideal)
````
